# Matplotlib lesson notebook

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Line charts

Limits - plt.xlim, plt.ylim  
Color - c argument  
Alpha - transparency  
Multiple plots    
plt.plot(x values first)

```python
x = np.array(range(1,11))

x
```

```python
y = x ** 2
y
```

```python
plt.plot(y)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlim(5,10)
```

```python
plt.plot(x,y)
plt.xlim(5,10)
plt.ylim(60,100)
```

```python
plt.plot(x,y, c= 'red')
plt.xlim(5,10)
plt.ylim(60,100)
```

```python
z = x ** 2 - 4
```

```python
plt.plot(x,y, c = 'oldlace')
plt.plot(x, z, c = 'firebrick', alpha = 1) #modding color and opacity
```

## Scatter charts

Size - s argument

```python
plt.scatter(x, y, s = 600)
plt.scatter(x, z)
plt.title('My blueberry $\pi$s')
plt.xlabel('My x values')
plt.ylabel('My y values')
```

## Custom Ticks

plt.xticks  
plt.yticks  

```python
plt.scatter(x, y, s = 600)
plt.scatter(x, z)
plt.xticks([2,4,8])
plt.yticks([33,45,77])
```

## Custom Line Types

Line style - ls argument (accepts ":" and "- -")

```python
plt.plot(x, y, ls = ':')
plt.plot(x, z, ls = '--')
```

## Rotating text and annotations

Rotation - rotation argument

```python
plt.plot(x, y, ls = ':')
plt.plot(x, z, ls = '--')
plt.xticks(rotation = 45)
```

## Adding text

plt.text - x and y coordinates, text, fontsize, color

plt.annotate - text, pointer coordinates, text coordinates, arrow properties

```python
plt.plot(x, y, ls = ':')
plt.plot(x, z, ls = '--')
plt.text(2, 10, 'test', fontsize = 9, c = 'firebrick')
```

```python
plt.plot(x, y, ls = ':')
plt.plot(x, z, ls = '--')
plt.annotate('arrow', xy =(2, 10), xytext= (6,60), arrowprops={'width': 3})
```

## Figure size and legends

plt.figure - figsize argument to determine size of plot  
plt.legend - loc argument for location

```python
plt.figure(figsize=(4,8), dpi = 20)
plt.plot(x, y, ls = ':')
plt.plot(x, z, ls = '--')
```

```python

plt.plot(x, y, ls = ':')
plt.plot(x, z, ls = '--')
plt.legend([1,2])
```

## Subplots

plt.subplot - number of rows, number of columns, selected figure  
plt.subplot - shorthand notation  
plt.suptitle - title for collection of subplots

```python
plt.subplot(1, 2, 1)
plt.plot(x, y)

plt.subplot(1,2,2)
plt.plot(x,z, c='firebrick')

plt.suptitle('A Tale of Two graphs')
```
